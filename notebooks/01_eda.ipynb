# Exploratory Data Analysis
This notebook will be used for EDA on the Heart Disease dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.loader import load_heart_disease_data

# Load dataset
df = load_heart_disease_data()
print('Dataset shape:', df.shape)
print(df.head())
print('\nDataset info:')
print(df.info())

# Missing value analysis
missing = df.isnull().sum()
print('\nMissing values per column:')
print(missing[missing > 0])

# Basic summary statistics
print('\nDescriptive statistics:')
print(df.describe(include='all').T)

# Class distribution
target_col = 'target' if 'target' in df.columns else 'num'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x=target_col)
plt.title('Class Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()

# Correlation heatmap
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Feature relationship example
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x=target_col, y='thalach')
plt.title('Maximum Heart Rate by Target Class')
plt.xlabel('Target')
plt.ylabel('Maximum Heart Rate')
plt.show()

# Note for the report
print('EDA completed. The notebook highlights class balance, missing values, correlations, and feature relationships.')
```
